```python
filename_weather = './data/weather.csv'
df_weather = pd.read_csv(filename_weather)
```

```python
df_weather.head()
```

```python
df_weather_long = pd.melt(
    df_weather,
    id_vars=['id', 'year', 'month', 'element'],
    var_name = 'day',
    value_name = 'temp'
)
```

```python
df_weather_long.head()
```

```python
df_weather_tidy = pd.pivot_table(
    df_weather_long,
    index = ['id', 'year', 'month', 'day'],
    columns = 'element',
    values = 'temp'
)
```

```python
df_weather_tidy
```

```python
# Flatten the DataFrame
df_weather_tidy = df_weather_tidy.reset_index()
```
